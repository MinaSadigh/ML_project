```python
# This code is a slightly modied version of the code available online at https://github.com/logpai/logparser/blob/master/demo/Spell_demo.py
import Spell
input_dir  = 'OpenStack/'
output_dir = 'OpenStack/ParsingResults'
log_file   = 'openstack_normal2.log'
log_format = '<Linerecord> <Date> <Time> <Pid> <Level> <Component> \[<ADDR>\] <Content>'
tau        = 0.5
regex      = []

parser = Spell.LogParser(indir=input_dir, outdir=output_dir, log_format=log_format, tau=tau, rex = regex, keep_para = False)
parser.parse(log_file)
```
